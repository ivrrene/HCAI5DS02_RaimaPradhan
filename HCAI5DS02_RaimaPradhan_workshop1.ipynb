```python
import numpy as np
# Simulating GPA data
np.random.seed(42)

dropout_gpa = np.random.normal(loc=2.1, scale=0.6, size=30)
retained_gpa = np.random.normal(loc=3.1, scale=0.5, size=90)

#Calculation Mean and SD
mean_dropout = np.mean(dropout_gpa)
mean_retained = np.mean(retained_gpa)
std_dropout = np.std(dropout_gpa, ddof=1)
std_retained = np.std(retained_gpa, ddof=1)

# Calculating Coefficient of Variation (CV)
cv_dropout = (np.std(dropout_gpa, ddof=1) / np.mean(dropout_gpa)) * 100
cv_retained = (np.std(retained_gpa, ddof=1) / np.mean(retained_gpa)) * 100

print(f"Droput Mean:{mean_dropout:.2f}")
print(f"Dropout SD:{std_dropout:.2f}")
print(f"Dropout CV: {cv_dropout:.2f}%\n")

print(f"Retained Mean:{mean_retained:.2f}")
print(f"Retained SD:{std_retained:.2f}")
print(f"Retained CV: {cv_retained:.2f}%")
```

```python
# Removing 5 random GPA values from dropout group
dropout_missing = np.delete(dropout_gpa, np.random.choice(len(dropout_gpa), 5, replace=False))

# Recalculating stats
mean_missing = np.mean(dropout_missing)
std_missing = np.std(dropout_missing, ddof=1)
cv_missing = (std_missing / mean_missing) * 100

print(f"New Mean (after removing 5 GPAs): {mean_missing:.2f}")
print(f"New Std Dev: {std_missing:.2f}")
print(f"New CV: {cv_missing:.2f}%")
```

```python
import matplotlib.pyplot as plt

data = [dropout_gpa, retained_gpa]
labels = ['Dropped Out', 'Retained']

plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=labels, patch_artist=True,
            boxprops=dict(facecolor='skyblue'),
            medianprops=dict(color='red', linewidth=2))

plt.title('GPA Boxplot: Dropped Out vs Retained')
plt.ylabel('GPA')
plt.grid(True)
plt.show()
```

Compared GPA variation between both groups using numerical summary.
Plotted boxplots to visually assess GPA distribution and spread.

Conclusion: Dropout students had a lower average GPA with more variability, while retained students had higher and more consistent GPAs. The higher CV for dropouts supported this.
```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate salary data (in $1000s)
male_salaries = np.random.normal(loc=105, scale=18, size=100)
female_salaries = np.random.normal(loc=92, scale=25, size=80)

# Function to print summary stats
def summary_stats(data, label):
    mean = np.mean(data)
    median = np.median(data)
    std = np.std(data, ddof=1)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1

    print(f"\n{label}")
    print("Mean:", round(mean, 2))
    print("Median:", round(median, 2))
    print("Standard Deviation:", round(std, 2))
    print("Q1:", round(q1, 2))
    print("Q3:", round(q3, 2))
    print("IQR:", round(iqr, 2))

# Print summary for both genders
summary_stats(male_salaries, "Male Salaries")
summary_stats(female_salaries, "Female Salaries")
```

```python
# Boxplot visualization
plt.figure(figsize=(8, 5))
plt.boxplot([female_salaries, male_salaries], tick_labels=["Female", "Male"], patch_artist=True, boxprops=dict(facecolor='skyblue'))
plt.title("Salary Distribution by Gender")
plt.ylabel("Salary ($1000s)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

Calculated mean, median, standard deviation, IQR, and CV for both genders.

Visualized salary distributions using boxplots.
Observed:

*  Mean > median â†’ indicates right-skewness for both groups

*  Females showed higher SD and CV â†’ more salary variability

*  Median is a more robust measure due to outliers
Noted: Departmental breakdown (Simpsonâ€™s Paradox) could change the interpretation if males dominate high-paying departments.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate 30 days of cart values
np.random.seed(1)
new_users = np.random.normal(45, 20, 30)       # Higher variability
returning_users = np.random.normal(60, 10, 30) # More consistent

# Clip any negative cart values (not realistic)
new_users = np.clip(new_users, 0, None)
returning_users = np.clip(returning_users, 0, None)

cv_new = np.std(new_users) / np.mean(new_users)
cv_returning = np.std(returning_users) / np.mean(returning_users)

print(f"CV - New Users: {cv_new:.2f}")
print(f"CV - Returning Users: {cv_returning:.2f}")
```

Here, since our new users are greater than returning users, new users are less predictable in spending.

```python
from scipy.stats import skew

skew_new = skew(new_users)
skew_returning = skew(returning_users)

print(f"Skewness - New Users: {skew_new:.2f}")
print(f"Skewness - Returning Users: {skew_returning:.2f}")
```

Positive skew: Long right tail (a few high-spending days).

New Users often show more skew due to promotions or inconsistent behavior.

Business implication: Median might be a better central measure than mean.

```python
# Simulate campaign effect
new_users[5] += 100
new_users[12] += 150

# Recalculate mean & median
print("Mean after campaign:", np.mean(new_users))
print("Median after campaign:", np.median(new_users))
```

Mean increases sharply due to outliers.

Median is more stable â†’ robust against extreme values.

Business reporting should highlight both to avoid misinterpretation.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read CSV
df = pd.read_csv("/content/drive/MyDrive/daily_sales.csv")
df['date'] = pd.to_datetime(df['date'])

# Split by Category
cat_a = df[df['category'] == 'Category A']
cat_b = df[df['category'] == 'Category B']

# Boxplots & Histograms
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.boxplot(data=df, x='category', y='total_sales')
plt.title('Boxplot of Total Sales')

plt.subplot(2, 2, 2)
sns.histplot(cat_a['total_sales'], kde=True, color='skyblue', label='Category A')
sns.histplot(cat_b['total_sales'], kde=True, color='salmon', label='Category B')
plt.title('Histogram of Total Sales')
plt.legend()

plt.subplot(2, 2, 3)
sns.boxplot(data=df, x='category', y='num_transactions')
plt.title('Boxplot of Number of Transactions')

plt.subplot(2, 2, 4)
sns.boxplot(data=df, x='category', y='avg_basket')
plt.title('Boxplot of Average Basket Size')

plt.tight_layout()
plt.show()

# Time Series Plot
plt.figure(figsize=(14, 6))
sns.lineplot(data=df, x='date', y='total_sales', hue='category', marker='o')
plt.title('Time Series of Daily Total Sales')
plt.xlabel('Date')
plt.ylabel('Total Sales (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Coefficient of Variation
def coefficient_of_variation(series):
    return series.std() / series.mean()

cv_results = pd.DataFrame({
    'Category': ['Category A', 'Category B'],
    'CV_Total_Sales': [coefficient_of_variation(cat_a['total_sales']),
                       coefficient_of_variation(cat_b['total_sales'])],
    'CV_Avg_Basket': [coefficient_of_variation(cat_a['avg_basket']),
                      coefficient_of_variation(cat_b['avg_basket'])]
})

print("Coefficient of Variation:")
print(cv_results)

#7-Day Moving Average Plot
df_sorted = df.sort_values('date')
moving_avg = df_sorted.groupby('category').apply(lambda x: x.set_index('date')[['total_sales']]
                                                 .rolling(7).mean()).reset_index()

plt.figure(figsize=(14, 6))
sns.lineplot(data=moving_avg, x='date', y='total_sales', hue='category')
plt.title('7-Day Moving Average of Total Sales')
plt.xlabel('Date')
plt.ylabel('7-Day Moving Avg (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Category A clearly has a higher and more consistent number of transactions, as expected from lower-cost, high-volume items.

Category B shows a lower median and more spread out valuesâ€”typical for premium products with fewer buyers.

3. Average Basket Size
Category B has a significantly higher average basket size (around $90â€“110), while category A stays below 30.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load Data
df = pd.read_csv("/content/drive/MyDrive/customer_segments.csv")
df['date'] = pd.to_datetime(df['date'])

# Violin Plot & Strip Plot
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='segment', y='transaction_amount', inner='box', palette='Pastel1')
sns.stripplot(data=df, x='segment', y='transaction_amount', color='black', size=3, alpha=0.5)
plt.title("Violin and Strip Plot of Transaction Amount by Segment")
plt.show()

# KDE (Density) Plot
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df, x='transaction_amount', hue='segment', fill=True, common_norm=False, palette='Set2')
plt.title("Density Plot of Transaction Amount by Segment")
plt.xlabel("Transaction Amount")
plt.ylabel("Density")
plt.show()

# Summary Statistics
summary = df.groupby('segment').agg(
    mean_spend=('transaction_amount', 'mean'),
    median_spend=('transaction_amount', 'median'),
    iqr_spend=('transaction_amount', lambda x: x.quantile(0.75) - x.quantile(0.25)),
    std_spend=('transaction_amount', 'std'),
    count=('transaction_amount', 'count')
)
summary['cv_spend'] = summary['std_spend'] / summary['mean_spend']
summary = summary.drop(columns='std_spend')
print("Summary Statistics:")
display(summary)

#  Boxplot + Swarmplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='segment', y='transaction_amount', palette='Set3')
sns.swarmplot(data=df, x='segment', y='transaction_amount', color='black', size=3, alpha=0.6)
plt.title("Boxplot with Swarm Overlay by Segment")
plt.show()

# Scatter Plot â€“ # Transactions vs Total Spend per Customer
customer_stats = df.groupby(['customer_id', 'segment']).agg(
    total_spent=('transaction_amount', 'sum'),
    num_transactions=('transaction_amount', 'count')
).reset_index()

plt.figure(figsize=(12, 6))
sns.scatterplot(data=customer_stats, x='num_transactions', y='total_spent', hue='segment', palette='Dark2')
plt.title("Total Spending vs Number of Transactions by Customer")
plt.xlabel("Number of Transactions")
plt.ylabel("Total Spending")
plt.grid(True)
plt.show()
```
