**2. Vectorization vs Loops**

• Write a function to compute element-wise square of an array using:
– a for-loop
– NumPy vectorized operation


```python
def square_with_loop(arr):
    result = np.empty_like(arr)
    for i in range(len(arr)):
        result[i] = arr[i] ** 2
    return result
def square_with_vector(arr):
    return arr ** 2
arr = np.array([1, 2, 3, 4, 5])

print("Using for-loop:", square_with_loop(arr))
print("Using vectorized:", square_with_vector(arr))
```

• Compare their execution time using %%timeit or time module.

```python
import numpy as np
import time

arr = np.arange(1000000)

%timeit arr ** 2

def square_with_loop(arr):
    result = np.empty_like(arr)
    for i in range(len(arr)):
        result[i] = arr[i] ** 2
    return result

%timeit square_with_loop(arr)
```

**Problem 4 – Subplots**

1. Create a 2x2 subplot grid showing:
• Line plot of sine
• Histogram of income
• Bar chart of region-wise population
• Boxplot of house value grouped by age group

```python
sns.set(style='whitegrid')
# Create subplot grid
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Line plot of sine
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
axes[0, 0].plot(x, y, color='blue')
axes[0, 0].set_title("Line Plot of y = sin(x)")
axes[0, 0].set_xlabel("x")
axes[0, 0].set_ylabel("sin(x)")

# 2. Histogram of income
axes[0, 1].hist(df['MedInc'], bins=30, color='orange', edgecolor='black')
axes[0, 1].set_title("Histogram of Median Income")
axes[0, 1].set_xlabel("MedInc")
axes[0, 1].set_ylabel("Frequency")

# 3. Bar chart of population by region
pop_by_region = df.groupby('Region')['Population'].sum()
axes[1, 0].bar(pop_by_region.index, pop_by_region.values, color='green', edgecolor='black')
axes[1, 0].set_title("Total Population by Region")
axes[1, 0].set_xlabel("Region")
axes[1, 0].set_ylabel("Total Population")

# 4. Boxplot of MedHouseVal by AgeGroup
sns.boxplot(data=df, x='AgeGroup', y='MedHouseVal', ax=axes[1, 1], palette='pastel')
axes[1, 1].set_title("House Value by Age Group")
axes[1, 1].set_xlabel("Age Group")
axes[1, 1].set_ylabel("Median House Value")

# Adjust layout
plt.tight_layout()
plt.show()
```
